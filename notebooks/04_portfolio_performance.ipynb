# Portfolio Performance Evaluation

This notebook introduces basic portfolio performance metrics and compares two portfolios using risk-adjusted indicators.

```python
import numpy as np
```

## Sample Portfolio Returns

We define two hypothetical portfolios and their monthly returns.

```python
portfolio_A = np.array([0.02, 0.01, -0.01, 0.03, 0.015, 0.005, 0.01, 0.02, -0.005, 0.01, 0.015, 0.025])
portfolio_B = np.array([0.03, -0.01, 0.00, 0.04, 0.02, -0.005, 0.015, 0.03, -0.01, 0.005, 0.02, 0.03])

risk_free_rate = 0.002
```

## Average Return and Volatility

We calculate the mean return and standard deviation for each portfolio.

```python
mean_A = np.mean(portfolio_A)
vol_A = np.std(portfolio_A)

mean_B = np.mean(portfolio_B)
vol_B = np.std(portfolio_B)

print("Portfolio A - Mean Return:", round(mean_A, 4))
print("Portfolio A - Volatility:", round(vol_A, 4))
print("Portfolio B - Mean Return:", round(mean_B, 4))
print("Portfolio B - Volatility:", round(vol_B, 4))
```

## Sharpe Ratio

The Sharpe ratio measures excess return per unit of risk.

```python
sharpe_A = (mean_A - risk_free_rate) / vol_A
sharpe_B = (mean_B - risk_free_rate) / vol_B

print("Portfolio A - Sharpe Ratio:", round(sharpe_A, 4))
print("Portfolio B - Sharpe Ratio:", round(sharpe_B, 4))
```

## Interpretation

A higher Sharpe ratio indicates better risk-adjusted performance.

```python
if sharpe_A > sharpe_B:
    print("Portfolio A has the better risk-adjusted performance.")
else:
    print("Portfolio B has the better risk-adjusted performance.")
```

## Simple Active Management Perspective

A portfolio manager is not only interested in returns, but also in how much risk was taken to achieve those returns.

```python
import matplotlib.pyplot as plt

labels = ["Portfolio A", "Portfolio B"]
mean_returns = [mean_A, mean_B]
volatilities = [vol_A, vol_B]

x = np.arange(len(labels))
width = 0.35

plt.bar(x - width/2, mean_returns, width, label="Mean Return")
plt.bar(x + width/2, volatilities, width, label="Volatility")

plt.xticks(x, labels)
plt.title("Portfolio Return and Risk Comparison")
plt.legend()
plt.show()
```
