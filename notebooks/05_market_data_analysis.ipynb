# Market Data Analysis

This notebook demonstrates how financial market data can be downloaded and analyzed using Python.

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

## Download Market Data

We download historical price data for Apple (AAPL).

```python
data = yf.download("AAPL", start="2020-01-01", end="2024-01-01")

data.head()
```

## Price Chart

```python
data["Close"].plot(title="Apple Stock Price")
plt.show()
```

## Daily Returns

```python
returns = data["Close"].pct_change().dropna()

returns.head()
```

```python
returns.hist(bins=50)
plt.title("Distribution of Returns")
plt.show()
```

## Return Statistics

```python
mean_return = returns.mean()
volatility = returns.std()

print("Average daily return:", round(mean_return,5))
print("Daily volatility:", round(volatility,5))
```
